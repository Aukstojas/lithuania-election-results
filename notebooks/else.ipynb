```python
from typing import Union, List
def contains_phrases(href: str, phrase) -> bool:
    """
    Check if the href contains all the specified phrases.

    Parameters:
        href (str): The string to search within.
        phrases (str or list of str): The phrase(s) to search for.

    Returns:
        bool: True if all phrases are found in href, False otherwise.
    """
    if isinstance(phrase, list):
        return all(p in href for p in phrase)
    elif isinstance(phrase, str):
        return phrase in href
    else:
        raise TypeError("phrases must be a string or a list of strings")
```

```python
def prepare_data(df):
    # Ensure numeric columns are correctly typed
    df['total_apylinke_votes'] = pd.to_numeric(df['total_apylinke_votes'], errors='coerce')
    df['total_pirm_votes'] = pd.to_numeric(df['total_pirm_votes'], errors='coerce')
    df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

    # Drop rows where total_apylinke_votes == 0 or NaN, since rate = p / v would be invalid
    df = df.dropna(subset=['total_apylinke_votes', 'rate'])
    df = df[df['total_apylinke_votes'] > 0]

    return df
```

```python
def model_2_weighted_zscore(df):
    # Group by party
    grp = df.groupby('party_name')['rate']
    mean_rate = grp.transform('mean')
    # For an approximation of sigma_party, we could just do a single std as in Model 1
    # or re-estimate it ignoring v. We'll keep it simple:
    std_rate = grp.transform('std')

    # Weighted z = (r - mean) / (std / sqrt(v))
    df['zscore_weighted'] = (df['rate'] - mean_rate) / (std_rate / np.sqrt(df['total_apylinke_votes']))

    # Mark suspect if above threshold
    df['suspect_weighted'] = df['zscore_weighted'] > 3  # or your chosen cutoff
    return df
```

```python


def model_3_mixed_effects(df):
   """
   We fit a simple random-intercept model:
      rate ~ 1 + C(party_name) + (1|apylinke)
   meaning:
      - Each party has its own fixed intercept (i.e., mean).
      - Each precinct gets a random intercept capturing baseline difference.
   """
   # statsmodels MixedLM syntax:
   # MixedLM.from_formula("response ~ fixed_effects", groups=..., data=...)
   # We'll treat 'apylinke' as the grouping variable for random intercepts.

   # We'll keep only relevant columns to avoid formula confusion
   sub_df = df[['rate','apylinke','party_name']].dropna()

   # Convert party_name to a categorical if not already
   sub_df['party_name'] = sub_df['party_name'].astype('category')

   formula = "rate ~ C(party_name)"
   model = smf.mixedlm(formula, data=sub_df, groups=sub_df['apylinke'])
   mixedf = model.fit(method='nm', maxiter=200, disp=False)  # 'nm' = Nelder-Mead, can try other solvers
   print(mixedf.summary())

   # Predictions / residuals
   sub_df['fitted'] = mixedf.fittedvalues
   sub_df['resid'] = sub_df['rate'] - sub_df['fitted']

   # Merge back to main df (optional)
   df = df.merge(
      sub_df[['rate','apylinke','party_name','fitted','resid']],
      on=['apylinke','party_name','rate'],
      how='left',
      suffixes=('_orig', '_mixed')
   )
   return df, mixedf

df, mixedf = model_3_mixed_effects(df)

# In statsmodels, the random effects are in .random_effects
random_effects = mixedf.random_effects

# random_effects is a dict: {precinct_identifier: [intercept_value], ...}
outliers = {pid: val[0] for pid,val in random_effects.items() if val[0] > 1.00}  # threshold for suspicious

for precinct_id, re_value in outliers.items():
    print(f"Potentially suspicious: {precinct_id} random intercept = {re_value:.4f}")
```

```python
def model_4_lrt_approach(df, suspect_precincts=None):
    """
    If we suspect certain precincts might cheat, we add a
    precinct-specific binary indicator in the regression.
    Then do a likelihood-ratio test comparing the "cheating" model vs. the baseline model.
    """
    # 1) Fit baseline model (no cheat indicator)
    sub_df = df[['rate','apylinke','party_name']].copy()
    sub_df['party_name'] = sub_df['party_name'].astype('category')

    formula_null = "rate ~ C(party_name)"
    null_model = smf.mixedlm(formula_null, data=sub_df, groups=sub_df['apylinke'])
    fit_null = null_model.fit(method='nm', maxiter=200, disp=False)

    # 2) Construct a cheat_indicator for "suspect" precincts
    # For demonstration, let's say suspect precincts are from a set passed as argument
    if suspect_precincts is None:
        suspect_precincts = set()

    sub_df['cheat_indicator'] = sub_df['apylinke'].isin(suspect_precincts).astype(int)

    # 3) Fit alternative model with cheat_indicator as an additional fixed effect
    formula_alt = "rate ~ C(party_name) + cheat_indicator"
    alt_model = smf.mixedlm(formula_alt, data=sub_df, groups=sub_df['apylinke'])
    fit_alt = alt_model.fit(method='nm', maxiter=200, disp=False)

    # 4) Compare log-likelihoods
    ll_null = fit_null.llf
    ll_alt = fit_alt.llf
    lr_stat = 2 * (ll_alt - ll_null)

    # degrees of freedom = 1 if we only added one cheat_indicator
    # or # of suspect precincts if we have multiple separate indicators
    # but let's do the simplest assumption: dof=1
    from scipy.stats import chi2
    p_value = 1 - chi2.cdf(lr_stat, df=1)

    print("== LRT for Cheating Model ==")
    print(f"Null model logLik = {ll_null:.2f}")
    print(f"Alt model logLik  = {ll_alt:.2f}")
    print(f"LR stat           = {lr_stat:.2f}")
    print(f"p-value           = {p_value:.4f}")
    print("\nNull Model:\n", fit_null.summary())
    print("\nAlt Model:\n", fit_alt.summary())

    return fit_null, fit_alt, p_value

```

```python
import pandas as pd
import statsmodels.formula.api as smf

df = catalog.load("all_total_pirm_votes_by_apylinke")
df['rate'] = df['total_pirm_votes'] / df['total_apylinke_votes']

def two_stage_party_apylinke_effects(df):
    """
    1) Fit baseline model: rate ~ C(party_name) + C(apygarda).
    2) Compute residuals.
    3) Fit second model on residuals: resid ~ C(party_name):C(apylinke).
    4) Output the model results, including top 50 terms with lowest p-values.
    """
    #-------------------------------------
    # 1) Baseline model
    #-------------------------------------
    # Convert to categorical if not already
    df['party_name'] = df['party_name'].astype('category')
    df['apygarda']   = df['apygarda'].astype('category')
    df['apylinke']   = df['apylinke'].astype('category')

    # OLS ~ Maximum Likelihood under normal errors
    formula_baseline = "rate ~ C(party_name) + C(apygarda)"
    model_baseline = smf.ols(formula_baseline, data=df).fit()

    print("\n=== Baseline Model ===")
    print(model_baseline.summary())

    # Store baseline residuals
    df['resid_baseline'] = model_baseline.resid

    #-------------------------------------
    # 2) Second model on residuals
    #-------------------------------------
    # We want: resid_baseline ~ C(party_name):C(apylinke)
    # The interaction "C(party_name):C(apylinke)" means:
    # a separate coefficient for each (party_name, apylinke) combination.
    # This is akin to d_{party,apylinke} in your description.
    formula_second = "resid_baseline ~ C(party_name):C(apylinke)"
    model_second = smf.ols(formula_second, data=df).fit()

    print("\n=== Second Model (residual ~ party : apylinke) ===")
    print(model_second.summary())

    #-------------------------------------
    # 3) Identify 'd' terms and their p-values
    #-------------------------------------
    # The second model has many coefficients named like "C(party_name)[X]:C(apylinke)[Y]"
    # Let's extract them along with p-values into a DataFrame.
    params = model_second.params
    pvals  = model_second.pvalues

    # Combine into a single DataFrame
    results = pd.DataFrame({
        'coefficient': params.index,
        'estimate': params.values,
        'pvalue': pvals.values
    })

    # Filter to keep only interaction terms (since the intercept or other terms might appear)
    # Typically they look like "C(party_name)[XYZ]:C(apylinke)[123]" 
    mask_interaction = results['coefficient'].str.contains('C(party_name)') & \
                       results['coefficient'].str.contains('C(apylinke)')
    interaction_df = results[mask_interaction].copy()

    # Sort by p-value ascending
    interaction_df.sort_values('pvalue', inplace=True)

    # Grab the top 50
    top_50 = interaction_df.head(50)

    print("\n=== Top 50 (party:apylinke) terms with lowest p-values ===")
    print(top_50)

    return model_baseline, model_second, top_50

# ========================
# Example usage:
# ========================
# Suppose `df` already has columns: ['rate', 'party_name', 'apygarda', 'apylinke', ...]
model_baseline, model_second, top_50 = two_stage_party_apylinke_effects(df)
```

```python
import pandas as pd
import numpy as np

def find_most_unlikely_right_skewed(df, weight_col='total_apylinke_votes', 
                                    target_col='rate', group_col='party_name', top_n=10):
    """
    Identifies the top_n most unlikely observations based on a right-skewed distribution.
    
    Parameters:
    - df (pd.DataFrame): The prepared DataFrame.
    - weight_col (str): Column name for weights.
    - target_col (str): Column name for the target variable.
    - group_col (str): Column name for grouping (e.g., party).
    - top_n (int): Number of top unlikely observations to return.
    
    Returns:
    - pd.DataFrame: Top_n most unlikely observations.
    """
    # 1. Calculate weighted mean per group
    weighted_means = df.groupby(group_col).apply(
        lambda x: np.average(x[target_col], weights=x[weight_col])
    ).rename('weighted_mean')
    
    # 2. Merge weighted means back to the original DataFrame
    df = df.merge(weighted_means, on=group_col)
    
    # 3. Calculate residuals
    df['residual'] = df[target_col] - df['weighted_mean']
    
    # 4. Identify top_n observations with the highest residuals
    top_unlikely = df.nlargest(top_n, 'residual')
    
    return top_unlikely[['apygarda', 'apylinke', 'party_name', 'total_apylinke_votes', 'rate', 'weighted_mean', 'residual']]

# Example Usage:
# Assuming 'df' is your prepared DataFrame
most_unlikely = find_most_unlikely_right_skewed(df)
most_unlikely
```
